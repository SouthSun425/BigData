11. 인덱싱과 슬라이싱


```python
# 인덱싱
listbox = [2, 4, 6, 8, 10]
print(listbox[0]) # 첫번째 값
print(listbox[3]) # 네번째 값
print(listbox[-1]) # 마지막 값
```

```python
# 슬라이싱
print(listbox[0:3]) # 마지막 값 앞까지 출력한다 !!!!!!!!!
```

```python
# listbox[시작:끝:간격] (생략하면 [처음부터:끝까지:1])
print(listbox[1:3])
print(listbox[:3])
print(listbox[3:])
print(listbox[::2])
```

```python
# 값 변경(슬라이싱 활용)
listbox[1:3] = [10, 20]
print(listbox)
```

13. 문자열

```python
# 문자 변경 replace(변경 전 단어, 변경 후 단어)
text = "빅데이터 분석기사 파이썬 공부"
text = text.replace("공부", "스터디")
print(text)

# 복수 변경 간단한거
text = text.replace("파이썬", "머신러닝").replace("분석기사", "분석을 위한")
print(text)
```

```python
# 문자열 분리 split()
print(date.split('-'))
print(text.split())
```

```python
# 문자열을 리스트로 반환하면 한글자씩 들어감
print(list(date))
```

14. 반복문


```python
# for 함수 - for 변수명 in 리스트 :
listbox = [2, 4, 6, 8, 10]
for item in listbox :
  print(item)

listbox = ['길벗', '시나공', '빅분기']
for item in listbox :
  print(item)
```

```python
# range() 활용 range(N) 0 부터 N - 1개 까지 N 번 반복
for item in range(5) :
  print(item)

for i in range(5, 10) :
  print(i)
```

```python
# list.append()
listbox = []
for i in range(1, 6) :
  listbox.append(i)
print(listbox)
```

```python
# enumerate() 인덱스 번호와 값을 같이 출력
listbox = ['길벗', '시나공', '빅분기', '분석']
for index, item in enumerate(listbox) :
  print(index, item)
```

```python
# zip() 여러개 값을 묶어준다 딕셔너리 키, 값 출력
person_info = {
    'name' : '사랑',
    'age' : 20,
    'city' : '부산',
    'hobbies' : ['연애', '수영', '코딩']
}

for k, v in zip(person_info.keys(), person_info.values()) :
  print(k, v)
```

15. 함수

```python
# 일반 함수
def hello() :
  print("안녕하세요!")

hello()
```

```python
# 매개변수가 있는 함수
def hello(name) :
  print(name + "님 안녕하세요!!")

hello('태양')
```

```python
# 더하기 함수
def plus(x, y) :
  print(x + y)

a = 2
b = 3
plus(a, b)
```

```python
# 반환값이 있는 함수
def plus(x, y) :
  result = x + y
  return result

a = plus(2, 3)
print(a)
```

```python
# 복수 반환값
listbox = [15, 46, 78, 24, 56]

def min_max(data) :
  mi = min(data)
  ma = max(data)
  return mi, ma

a, b = min_max(listbox)
print(a, b)
```

```python
# 평균을 구하는 함수
listbox = [15, 46, 78, 24, 56]

def mean(data) :
  return sum(data) / len(data)

print(mean(listbox))
```
